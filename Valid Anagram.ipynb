### Problem: Is Anagram
Given two strings `s` and `t`, determine if `t` is an anagram of `s`.

- An anagram means both strings contain the **same characters** with the **same frequency**, but their order can differ.
  
**Example**:
```plaintext
Input: s = "anagram", t = "nagaram"
Output: true

Input: s = "rat", t = "car"
Output: false
```

---

### Approach 1: Sorting the Strings
Sort both strings and compare them.

#### **Code**
```python
def isAnagram(s, t):
    return sorted(s) == sorted(t)
```

#### **Explanation**
- Sorting ensures the characters are in the same order.
- If both strings match after sorting, they are anagrams.

#### **Complexity**
- **Time Complexity**: \(O(n \log n)\), due to sorting.
- **Space Complexity**: \(O(n)\), for storing the sorted strings.

---

### Approach 2: Using a Hash Map (Frequency Count)
Count the frequency of each character in both strings and compare the counts.

#### **Code**
```python
from collections import Counter

def isAnagram(s, t):
    return Counter(s) == Counter(t)
```

#### **Explanation**
- `Counter(s)` creates a frequency map for `s` (e.g., `{"a": 3, "n": 1, "g": 1, "r": 1, "m": 1}`).
- Compare the two frequency maps. If they match, `t` is an anagram of `s`.

#### **Complexity**
- **Time Complexity**: \(O(n)\), where \(n\) is the length of `s` or `t`.
- **Space Complexity**: \(O(1)\) for the map (only 26 English lowercase letters).

### Approach 3: Manual Frequency Count (Without `Counter`)
Manually count characters using an array (or dictionary).

#### **Code**
```python
def isAnagram(s, t):
    if len(s) != len(t):
        return False
    
    freq = [0] * 26  # For 26 lowercase English letters

    for char in s:
        freq[ord(char) - ord('a')] += 1
    
    for char in t:
        freq[ord(char) - ord('a')] -= 1
        if freq[ord(char) - ord('a')] < 0:
            return False

    return all(f == 0 for f in freq)
```

#### **Explanation**
- Use an array of size 26 (one for each letter).
- Increment counts for `s`, decrement for `t`.
- If any value in the array is not 0, the strings arenâ€™t anagrams.

#### **Complexity**
- **Time Complexity**: \(O(n)\).

### Approach 4: Two Hash Maps
Use two separate dictionaries to count character frequencies for `s` and `t`.

#### **Code**
```python
def isAnagram(s, t):
    if len(s) != len(t):
        return False
    
    freq_s = {}
    freq_t = {}

    for char in s:
        freq_s[char] = freq_s.get(char, 0) + 1
    for char in t:
        freq_t[char] = freq_t.get(char, 0) + 1

    return freq_s == freq_t
```

#### **Complexity**
- **Time Complexity**: \(O(n)\).
- **Space Complexity**: \(O(1)\) for fixed character set.

### Approach 5: Single Hash Map
Instead of two hash maps, use one map and adjust counts for `s` and `t`.

#### **Code**
```python
def isAnagram(s, t):
    if len(s) != len(t):
        return False
    
    freq = {}

    for char in s:
        freq[char] = freq.get(char, 0) + 1

    for char in t:
        if char not in freq or freq[char] == 0:
            return False
        freq[char] -= 1

    return all(value == 0 for value in freq.values())
```

#### **Complexity**
- **Time Complexity**: \(O(n)\).
- **Space Complexity**: \(O(1)\).
