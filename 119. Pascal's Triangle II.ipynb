```python
def getRow(rowIndex):
    if not rowIndex:
        return [1]
    result=[]
    for i in range(1,rowIndex+2):
        if i <= 2:
            result.append([1]*i)
        if i>=3:
            a,b=0,1
            result.append([1] * i)
            for j in range(1,len(result[i-2])):
                result[i-1][j]=result[i-2][a]+result[i-2][b]
                a+=1
                b+=1
    return result[rowIndex]

print(getRow(4))
```
